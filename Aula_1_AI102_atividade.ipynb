```python
vel = float(input("Digite a velocidade que voce passou no carro na avenida: "))
if vel > 60:
  calc = vel - 60
  multa = (10 * calc) + 200
  print("MULTADO! O valor da sua multa é R$ ", multa)
else:
  print("Velocidade dentro do limite. Boa viagem!")
```

```python
# Exemplo elif

cor = input("Digite a cor atual do semáforo (verde/amarelo/vermelho)")
if cor == "verde":
  print("Pode passar, caminho livre!!!")
elif cor == "amarelo":
  print("Tenha atenção ao cruzar o farol!!!")
elif cor == "vermelho":
  print("Caminho bloqueado, pare!!!")
else:
  print("Cor digitada inválida!!!")
```
