```python
# Exemplo elif

cor = input("Digite a cor atual do semáforo (verde/amarelo/vermelho)")
if cor == "verde":
  print("Pode passar, caminho livre!!!")
elif cor == "amarelo":
  print("Tenha atenção ao cruzar o farol!!!")
elif cor == "vermelho":
  print("Caminho bloqueado, pare!!!")
else:
  print("Cor digitada inválida!!!")
```

```python
##Exercicio 11

print("Programa que calcula o valor das moedas ganhas no jogo: ")
total_moedas = 0
fases = int(input("Digite a quantidade fases que terão no jogo "))

for i in range (1,fases + 1):
  for j in range (1,6):
    total_moedas = total_moedas + j;


print (f"O resultado da soma de moedas é R$ {total_moedas}")
```
